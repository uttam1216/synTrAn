```python
from haversine import haversine, Unit
import pickle
import os
import osmnx as ox
import networkx as nx
import pandas as pd
import osmnx as ox
import networkx as nx
import geohash as geohash
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import staticmaps
```

```python
allfiles=os.listdir('filtered_4749_traj/fil_traj_less_than_700_car')
len(allfiles)
```

```python
user_name=[]
latitude=[]
longitude=[]
date_time=[]
altitude_feet=[]
vehicle_type=[]
# for i in range(1):
for i in range(len(allfiles)):
    with open('filtered_4749_traj/fil_traj_less_than_700_car/'+allfiles[i], 'rb') as f:
        data = pickle.load(f)
        files_name=allfiles[i].split('_')
        user_name.append(files_name[0])
        vehicle_type.append(files_name[1])
        latitude.append(list(data['latitude']))
        longitude.append(list(data['longitude']))
        date_time.append(list(data['date_time']))
        altitude_feet.append(list(data['altitude_feet']))



    
```

```python
df = pd.DataFrame(list(zip(user_name, latitude,longitude,date_time,altitude_feet,vehicle_type)),
               columns =['user_name', 'latitude','longitude','date_time','altitude_feet','vehicle_type'])
```

```python
df
```

```python

User_infos=[]
trajectories=[]
time_stamps=[]
categories=[]
for i in range(df.shape[0]):
# for i in range(1):
    time_stamp=[]
    trajectory=[]
    category=[]
    for j in range(len(df['latitude'][i])-1):
        distance=haversine((df['latitude'][i][j],df['longitude'][i][j]), (df['latitude'][i][j+1],df['longitude'][i][j+1]),unit='km')
        if(distance!=0.0):
            padding=distance
            distance=round(distance,8)
            distance=str(distance).replace('.','')
            distance=str(distance).replace('e-','')
            distance=int(distance)
            time=int(df['date_time'][i][j].timestamp())
            time_stamp.append(time)
            trajectory.append(distance)
    
    
#     print(df['user_name'][i])
    if(df['vehicle_type'][i]=='taxi'):
        category.append(0)
    else:
        category.append(1)
    if(df['user_name'][i]<8):
        print(df['user_name'][i])
    if(len(trajectory)>8):
#         print(len(trajectory))
        User_infos.append(df['user_name'][i])
        trajectories.append(trajectory)
        time_stamps.append(time_stamp)   
        categories.append(category)   
        print("*******************************************************************************")
    
    
```

```python
main_df = pd.DataFrame(list(zip(User_infos, trajectories,time_stamps,categories)),
               columns =['user', 'traj','time','category'])
```

```python
main_df.to_csv('mainTUL_Geolife.csv', index=False,header=False)
```
